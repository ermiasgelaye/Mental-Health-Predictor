```python
# Exploratory Data Analysis
explore_survey = pd.read_csv('Data/survey.csv')
# changing "Yes" and "No" values by 1 and 0 (to make plots possible)

explore_survey.replace(('Yes', 'No'), (1, 0), inplace=True)

# Gender versus treatment
gender_treatment= explore_survey.pivot_table(index="Gender",values="treatment")
gender_treatment.plot.bar(figsize=(10,8),fontsize = 14)

plt.xlabel("Gender",fontsize = 14)
plt.ylabel("Proportion of people who sought treatment",fontsize = 14)
plt.title("Gender Versus Treatment",fontsize = 20)

plt.savefig("Image/gender_vs_treatment", bbox_inches = "tight")
plt.tight_layout()
plt.show()
```

```python
# Categorizing age to make ready for analysis
cut_points = [17,30,50,75]
label_names = ["Young Adult","Adult","Mid-Older Adult"]
explore_survey["Age_categories"] = pd.cut(survey["Age"],cut_points,labels=label_names)
```

```python
# plot treatmnet by age group
treatment_age = explore_survey.pivot_table(index="Age_categories",values='treatment')
treatment_age.plot.bar(figsize=(10,8),fontsize = 14)

plt.xlabel("Age Group",fontsize = 14)
plt.ylabel("Proportion of people who sought treatment",fontsize = 14)
plt.title("Age Versus Treatment",fontsize = 20)

plt.savefig("Image/age_vs_treatment", bbox_inches = "tight")
plt.tight_layout()
plt.show()
```

```python
explore_survey["Age"].describe()
```

```python
# Some values in the "Age" column are not realistic or completely no sense (like negative values or positive values above 150).
#Values not between 18 and 73 will be filled by the median value of the column.
explore_survey['Age'][explore_survey['Age']<18] = explore_survey['Age'].median()
explore_survey['Age'][explore_survey['Age']>73] = explore_survey['Age'].median()
```

```python
survey=survey.fillna({"self_employed":"no_data"})
```

```python
survey=survey.fillna({"work_interfere":"no_data"})


survey["work_interfere"] = survey["work_interfere"].replace({
    "no_data":"Never"
})
```

```python
survey=survey.fillna({"comments":"no_comment"})


survey["comments"] = survey["comments"].replace({
    "-":"no_comment"
})

survey["mapcomment"] = survey["comments"].map({
    "no_comment":"no_comment"
})

survey=survey.fillna({"mapcomment":"some_comment"})
survey['mapcomment'].value_counts().plot(figsize=(10,8),kind='bar')
```

## Normalizing the data for the ML Analysis 









Normalizing data for ML analysis. The following items are not used. "Timestamp", "Gender", "comments". -The following three-value items "Yes", "No", and "no_data" are represented by "1" and "0". "Treatment", "family_history", "remote_work", "tech_company", "obs_consequence", ãƒ» Except for the above, try all dummy variables once.

```python
survey_data=survey.drop("Timestamp",axis=1).drop("Gender",axis=1).drop("comments",axis=1)
col=survey_data.columns.tolist()
col.remove('treatment')
col.insert(0,"treatment")
survey_data=survey_data.loc[:,col]

for column in survey_data.columns:
    # For YES / NO data, replace with No: 0, Yes: 1.
    if column in ["treatment", "family_history", "remote_work", "tech_company", "obs_consequence", "mapcomment"]:
        survey_data[column] = survey_data[column].map({"No": 0, "Yes": 1, "no_comment": 0, "some_comment": 1})


# Generate dummy variables
survey_data=pd.get_dummies(survey_data)

display(survey_data.head())
display(survey_data.shape)
```

```python
survey_corr= pd.get_dummies(survey_data).corr()
survey_corr_treatment= survey_corr["treatment"]

print("Display items with correlation coefficient of 0.2 or more and -0.2 or less")

# Get values with a correlation coefficient of 0.2 or more and -0.2 or less
target_cc_value = 0.2

print("Positive correlation")
display(survey_corr_treatment[survey_corr_treatment >= target_cc_value])
print("Negative correlation")
display(survey_corr_treatment[survey_corr_treatment <= (-1*target_cc_value)])

df_indexs = survey_corr[(survey_corr["treatment"] >= target_cc_value) | (survey_corr["treatment"] <= (-1 * target_cc_value))].index

#print(survey_corr)

df_cc_target=survey_corr.loc[df_indexs,df_indexs]
plt.figure(figsize=(10, 8)) #heatmap size
sns.heatmap(df_cc_target, annot=True, linewidths=.5)
```

```python
g = sns.pairplot(df_cc_target)
g.map_upper(sns.scatterplot, s=100,color = 'darkred')
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=2)
```

# Machine Learning Analysis

## 1. Linear regression (+ confusion matrix )

```python
# prepare the data
y = survey_data["treatment"].values
X = survey_data.iloc[:,1:].values
X_columns = survey_data.iloc[:,1:].columns

#X_train = X
#y_train = y
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Identify using optimal parameters
regr = LinearRegression(fit_intercept=True)
regr.fit(X_train, y_train)

y_train_pred = regr.predict(X_train)
#print(y_train_pred)

# Let's make it possible to specify the threshold value of whether is 0 or 1.
threshhold=0.6

y_train_pred[y_train_pred > threshhold] =1
y_train_pred[y_train_pred <= threshhold] =0

df = pd.DataFrame({"Actual":y_train,"Predicted":y_train_pred})  
df = df.groupby(["Actual","Predicted"])[["Predicted"]].count()  
df.rename(columns={"Predicted":"num"},inplace=True)  
df.reset_index(inplace=True)  
df = df.pivot(index="Actual",columns="Predicted",values="num")  


# Let's look at the distribution of values when linear regression is performed
# Consider the following matrix where the classes are â€œTreatment" or "no treatmentâ€ 
display(df)
plt.figure(figsize=(10, 8)) 
sns.heatmap(df, annot=True, cmap='GnBu', fmt='4d')
plt.show()

### Accuracy
accuracy = accuracy_score(y_train, y_train_pred)
### Recall
recall = recall_score(y_train, y_train_pred)
### Precision
precision = precision_score(y_train, y_train_pred)
### F1
f1 = f1_score(y_train, y_train_pred)

print('Accuracyï¼š{0:.2f}%'.format(accuracy * 100))
print('Recallï¼š{0:.2f}%'.format(recall * 100))
print('Precisionï¼š{0:.2f}%'.format(precision * 100))
print('F1ï¼š{0:.2f}%'.format(f1 * 100))
```

```python
## Verification by test data

y_test_pred = regr.predict(X_test)

# Allows you to specify a threshold for treat to be 0 or 1.

y_test_pred[y_test_pred > threshhold] =1
y_test_pred[y_test_pred <= threshhold] =0


df = pd.DataFrame({"Actual":y_test,"Predicted":y_test_pred})  
df = df.groupby(["Actual","Predicted"])[["Predicted"]].count()  
df.rename(columns={"Predicted":"num"},inplace=True)  
df.reset_index(inplace=True)  
df = df.pivot(index="Actual",columns="Predicted",values="num")  


display(df)
plt.figure(figsize=(10, 8)) 
sns.heatmap(df, annot=True, cmap='GnBu', fmt='4d')
plt.show()

### Accuracy
accuracy = accuracy_score(y_test, y_test_pred)
### Recall
recall = recall_score(y_test, y_test_pred)
### Precision
precision = precision_score(y_test, y_test_pred)
### F1
f1 = f1_score(y_test, y_test_pred)

print('Accuracyï¼š{0:.2f}%'.format(accuracy * 100))
print('Recallï¼š{0:.2f}%'.format(recall * 100))
print('Precisionï¼š{0:.2f}%'.format(precision * 100))
print('F1ï¼š{0:.2f}%'.format(f1 * 100))
```

## 7. Natural Language Processing (NLP)

```python
# Python program to generate WordCloud 
comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in tqdm(explore_survey.comments[~explore_survey.comments.isnull()]):
      
  # typecaste each val to string 
  val = str(val) 

  # split the value 
  tokens = val.split() 
    
  # Converts each token into lowercase 
  for i in range(len(tokens)): 
      tokens[i] = tokens[i].lower() 
    
  comment_words += " ".join(tokens)+" "

wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black', 
                stopwords = stopwords, 
                min_font_size = 12).generate(comment_words) 

# plot the WordCloud image                        
plt.figure(figsize = (12, 12), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
