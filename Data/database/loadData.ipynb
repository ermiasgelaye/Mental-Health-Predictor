#### Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plot
from sqlalchemy import create_engine
from sqlalchemy import inspect
from config import pg_user, pg_password, db_name, pg_host  # Corrected import
import plotly.graph_objs as go
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv

load_dotenv()
```

#### Connecting to the database

```python
pg_user = os.getenv("PG_USER")
pg_password = os.getenv("PG_PASSWORD")
pg_host = os.getenv("PG_HOST")
pg_port = os.getenv("PG_PORT", "5432")
db_name = os.getenv("DB_NAME")


DATABASE_URL = f"postgresql://{pg_user}:{pg_password}@{pg_host}:{pg_port}/{db_name}?sslmode=require"
DATABASE_URL = DATABASE_URL.replace('postgres://', 'postgresql://', 1)
```

#### Checking table schema

```python
engine = create_engine(DATABASE_URL)
```

#### Adding mental health data to schema 

```python
mental_helath_survey = '../survey.csv'
mental_helath_survey= pd.read_csv(mental_helath_survey , index_col=0)
mental_helath_survey
```

```python
mental_helath_survey.to_sql(name='mental_health', con=engine, if_exists='append', index=True)
```
